```python
#Dependencies
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# list of tv show titles to query
tv_shows = ["Altered Carbon", "Grey's Anatomy", "This is Us", "The Flash",
            "Vikings", "Shameless", "Arrow", "Peaky Blinders", "Dirk Gently"]

# tv maze show search base url
base_url = "http://api.tvmaze.com/search/shows?q="

# set up lists to hold response data for name and rating
titles = []
ratings = []

url = base_url+ "Vikings"
url
show_response = requests.get(url).json()
#show_response_1 = show_response[0]["show"]["name"]
show_response
```

```python
# loop through tv show titles, make requests and parse
for show in tv_shows:
    url = base_url+show
    show_response = requests.get(url).json()
    name = show_response[0]["show"]["name"]
    rating = show_response[0]["show"]["rating"]["average"]
    titles.append(name)
    ratings.append(rating)

shows = {"title" : titles,
        "Rating" : ratings}
```

```python
# create dataframe
shows_df = pd.DataFrame(shows)

shows_df
```

```python
#plt.bar(shows_df['title'],shows_df['Rating'])

shows_df.plot('title','Rating',kind = 'bar',figsize=(10,5))
plt.show()
```
