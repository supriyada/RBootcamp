```python
# Dependencies
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
import pandas as pd
```

```python
# Load crime data set into pandas
crime_data = pd.read_csv("../Resources/crime_data.csv")
crime_data.head()
```

```python
# Generate a scatter plot of year versus violent crime rate
year = crime_data['year']
violent_crime_rate = crime_data['violent crime rate']
plt.scatter(year,violent_crime_rate)

plt.xlabel('Year')
plt.ylabel('Violent Crime Rate')
plt.show()
```

```python
# Perform a linear regression on year versus violent crime rate
year = crime_data['year']
violent_crime_rate = crime_data['violent crime rate']
(slope, intercept, rvalue, pvalue, stderr) = linregress(year, violent_crime_rate)

```

```python
# Create equation of line to calculate predicted violent crime rate
regress_values = year * slope + intercept
```

```python
# Plot the linear model on top of scatter plot 
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(year,violent_crime_rate)

plt.plot(year,regress_values,"--")
#plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Year')
plt.ylabel('Violent Crime Rate')
plt.show()
```

```python
# Repeat plotting scatter and linear model for year versus murder rate
year = crime_data['year']
murder_rate = crime_data.iloc[:,5]
(slope1, intercept1, rvalue1, pvalue1, stderr1) = linregress(year, murder_rate)
murder_rate_values = year * slope1 + intercept1
plt.scatter(year,murder_rate)

plt.plot(year,murder_rate_values,"r-")
plt.xlabel('Year')
plt.ylabel('Murder Rate')
plt.show()
```

```python
# Repeat plotting scatter and linear model for year versus aggravated assault
year = crime_data['year']
assault_rate = crime_data['aggravated assault rate']
(slope2, intercept2, rvalue2, pvalue2, stderr2) = linregress(year, assault_rate)
regress_values2 = year * slope2 + intercept2
plt.scatter(year,assault_rate)

plt.plot(year,regress_values2,"r-")
plt.xlabel('Year')
plt.ylabel('Assault Rate')
plt.show()
```

```python
# Generate a facet plot of all 3 figures
fig,(ax1,ax2,ax3)=plt.subplots(3,sharex=True)
ax1.scatter(year,violent_crime_rate)
ax1.plot(year,regress_values,"--")
ax1.set_ylabel("Violent crime")

ax2.scatter(year,murder_rate)
ax2.plot(year,murder_rate_values,"r-")

ax3.scatter(year,assault_rate)
ax3.plot(year,regress_values2,"r-")

plt.show()
```
