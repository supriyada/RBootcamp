```python
# Dependencies
import pandas as pd
import numpy as np
```

```python
# Save file path to variable
pokemon_csv = "Resources/Pokemon.csv"
```

```python
# Read with Pandas
pokemon_df = pd.read_csv(pokemon_csv)
pokemon_df.head()
```

```python
# Extract the following columns: ["Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]
pokemon_type_df = pokemon_df[["Type 1", "HP", "Attack", "Sp. Atk", "Sp. Def", "Speed"]]
pokemon_type_df.head()
```

```python
# Create a dataframe of the average stats for each type of pokemon.
pokemon_group = pokemon_type_df.groupby(['Type 1'])

create_df = pokemon_group.mean()
create_df.head()
```

```python
# Calculate the total power level of each type of pokemon by summing all of the stats together.
# Place the results into a new column.
create_df["Total Power"] = create_df.sum(axis=1)
create_df["Total Power"]
create_df.head()
```

```python
create_df["Two Power"] = create_df.iloc[:,:].sum()
create_df["Two Power"]
create_df.head()
```

```python
# Bonus: Sort the table by strongest type and export the resulting table to a new CSV.
strongest_pokemon = create_df.sort_values(["Total Power"], ascending = False)
strongest_pokemon.head()
strongest_pokemon.to_csv("output/pokemon_rankings.csv", index=True)
```
