```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
crypto_df = pd.merge(bitcoin_df,dash_df,on = "Date")
crypto_df.head()
```

```python
# Rename columns so that they are differentiated
crypto_df = crypto_df.rename(columns={"Open_x":"Bitcoin Open","High_x":"Bitcoin High","Low_x":"Bitcoin Low","Close_x":"Bitcoin Close",
                                     "Volume_x":"Bitcoin Volume","Market Cap_x":"Bitcoin Market cap"})

crypto_df = crypto_df.rename(columns={"Open_y":"Dash Open","High_y":"Dash High","Low_y":"Dash Low",
                                      "Close_y":"Dash Close","Volume_y":"Dash Volume","Market Cap_y":"Dash Market cap"})
crypto_df.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
alternative_df = pd.merge(
    bitcoin_df, dash_df, on="Date", suffixes=("_Bitcoin", "_Dash"))
alternative_df.head()
```

```python
# Collecting best open for Bitcoin and Dash
bitcoin_open = alternative_df["Open_Bitcoin"].max()
dash_open = alternative_df["Open_Dash"].max()
print(bitcoin_open)
print(dash_open)

# Collecting best close for Bitcoin and Dash
bitcoin_close = alternative_df["Close_Bitcoin"].max()
dash_close = alternative_df["Close_Dash"].max()
print(bitcoin_close)
print(dash_close)

# Collecting the total volume for Bitcoin and Dash, display in millions
bitcoin_volume = alternative_df["Volume_Bitcoin"].sum()/100000
dash_volume = alternative_df["Volume_Dash"].sum()/1000000
print(bitcoin_volume)
print(dash_volume)
```

```python
# Creating a summary DataFrame using above values
summary_df = pd.DataFrame({"bitcoin_open":[bitcoin_open],
                          "dash_open":[dash_open],
                          "bitcoin_close":[bitcoin_close],
                          "dash_close":[dash_close]})

summary_df
```
