```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create dummy data
from sklearn.datasets.samples_generator import make_blobs

X, _ = make_blobs(n_samples=500, centers=6,
                            cluster_std=0.70, random_state=0)
```

```python
# Plot and show scatter
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# Create a kmeans model using k = 12
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 12)
# Fit the model to the data
kmeans.fit(X)

# Use the data to predict the clusters
# save the predictions as `predicted_clusters`
predicted_value = kmeans.predict(X)
```

```python
# Plot the clusters
plt.scatter(X[:,0], X[:,1], c= predicted_value, s = 50, cmap = 'viridis')
```

```python
# Print the cluster centers and cluster labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```
