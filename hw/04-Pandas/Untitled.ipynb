```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school = pd.read_csv(school_data_to_load)
student = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data = pd.merge(student, school, on="school_name")

print(school.head())
print(student.head())
school_data.head()
```

```python
school_summary=school_data.loc[:,["school_name","type","size","budget"]].groupby("school_name").first()
school_summary["Per student Budget"] = school_summary["budget"]/school_summary["size"]

school_summary=pd.concat([school_summary,(school_data.groupby("school_name").agg({'math_score':'mean','reading_score':'mean'}))],axis = 1)

school_summary=school_summary.rename(columns={"math_score":"Average Math Score","reading_score":"Average Reading Score"})

sch_math_pass_percent = (school_data.loc[school_data["math_score"] >= 70,:].groupby("school_name")["math_score"].count()/school_data.groupby("school_name")["math_score"].count())*100
sch_read_pass_percent = (school_data.loc[school_data["reading_score"] >= 70,:].groupby("school_name")["reading_score"].count()/school_data.groupby("school_name")["reading_score"].count())*100

school_summary = pd.concat([school_summary,sch_math_pass_percent,sch_read_pass_percent],axis=1)
school_summary=school_summary.rename(columns={"math_score":"% Passing Maths","reading_score":"% Passing Reading"})

sch_overall_percent = (school_data.loc[(school_data["math_score"] >= 70)&(school_data["reading_score"] >= 70),:].groupby("school_name")["math_score"].count()/school_data.groupby("school_name")["math_score"].count())*100
school_summary = pd.concat([school_summary,sch_overall_percent],axis=1)
school_summary=school_summary.rename(columns={"math_score":"% of Overall Passing"})   

school_summary_df=pd.DataFrame(school_summary)
school_summary_df
```

```python
grade_9_series = pd.Series(student["math_score"]).where(student["grade"] == "9th").groupby(student["school_name"]).mean().reset_index(name='9th')
grade_10_series = pd.Series(student["math_score"]).where(student["grade"] == "10th").groupby(student["school_name"]).mean().reset_index(name='10th')
grade_11_series = pd.Series(student["math_score"]).where(student["grade"] == "11th").groupby(student["school_name"]).mean().reset_index(name='11th')
grade_12_series = pd.Series(student["math_score"]).where(student["grade"] == "12th").groupby(student["school_name"]).mean().reset_index(name='12th')

math_score_by_grade_df=pd.concat([grade_9_series,grade_10_series,grade_11_series,grade_12_series],axis=1)

math_score_by_grade_df=math_score_by_grade_df.set_index("school_name")
math_score_by_grade_df=math_score_by_grade_df.rename_axis("School Name")

math_score_by_grade_df["9th"]=math_score_by_grade_df["9th"].map("{:,.2f}".format)
math_score_by_grade_df["10th"]=math_score_by_grade_df["10th"].map("{:,.2f}".format)
math_score_by_grade_df["11th"]=math_score_by_grade_df["11th"].map("{:,.2f}".format)
math_score_by_grade_df["12th"]=math_score_by_grade_df["12th"].map("{:,.2f}".format)

math_score_by_grade_df
```
